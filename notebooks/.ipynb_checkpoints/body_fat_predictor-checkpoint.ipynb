```python
def plot_hist(df, column):
    plt.figure(figsize=(10,5)) # Plot size
    plt.style.use('seaborn-whitegrid') # Style

    # Building a histogram
    n, bins, patches = plt.hist(df[column], bins=90, facecolor='#2ab0ff', edgecolor='#e0e0e0', linewidth=0.5, alpha=0.7)

    # Changing properties of each bin
    n = n.astype('int') # n should be integer, otherwise it wouldn't work properly

    # Accessing each bin, changing color according to height
    for i in range(len(patches)):
        patches[i].set_facecolor(plt.cm.viridis(n[i]/max(n)))

    # Pick one bin
    # patches[47].set_fc('red') # Color
    # patches[47].set_alpha(1) # Opacity

    # Making annotation
    # plt.annotate('Something important here!', xy=(0.57, 175), xytext=(2, 130), fontsize=10, arrowprops={'width':0.4,'headwidth':7,'color':'#333333'})

    # Adding title and labels with custom font sizes
    plt.title(f'{column} Distribution', fontsize=12)
    plt.xlabel('Distribution', fontsize=10)
    plt.ylabel('Frequency', fontsize=10)
    plt.show()
```

The prediction result you get is unbelievably good, 95%!
However, as can be seen in the graph showing the importance of each variable in the prediction, density is the most important variable by far in comparison to the rest of the variables.
This, in the experimental setting, is not a problem, because the data exists and, if available, it is possible to predict the body fat index with 95 % accuracy. But the aim of a good project should be to be able to transfer the solution from the experimental setting to the real world, and this is not very feasible considering that the density variable represents the density determined by weighing the person under water. This model would not save any time or cost in obtaining the body fat index, as the more expensive procedure of weighing people underwater would still have to be carried out (and it is not at all quick).
This is why we are going to try to develop a model without this variable, as the rest of the variables can be obtained from a person in a matter of a few minutes.

```python
def plot_models_comparison(logs):
    x = []
    y = []
    z = []
    for log in logs:
        x.append(log['name'])
        y.append(log['score'])
        z.append(log['rmse'])

    # sns.barplot(x,y)
    # sns.barplot(x,z)

    plt.figure(figsize=(20,10))
    plt.subplot(2,2,1)
    sns.barplot(x,y)
    plt.title("Models and their accuracy")

    plt.subplot(2,2,2)
    sns.barplot(x,z)
    plt.title("Models and their rmse")

    plt.show()
```

## SHAP values

```python
# Fits the explainer
explainer = shap.Explainer(lasso_regr.predict, X_test)
# Calculates the SHAP values - It takes some time
shap_values = explainer(X_test)
```

```python
def print_feature_importances_shap_values(shap_values, features):
    '''
    Prints the feature importances based on SHAP values in an ordered way
    shap_values -> The SHAP values calculated from a shap.Explainer object
    features -> The name of the features, on the order presented to the explainer
    '''
    # Calculates the feature importance (mean absolute shap value) for each feature
    importances = []
    for i in range(shap_values.values.shape[1]):
        importances.append(np.mean(np.abs(shap_values.values[:, i])))
    # Calculates the normalized version
    importances_norm = softmax(importances)
    # Organize the importances and columns in a dictionary
    feature_importances = {fea: imp for imp, fea in zip(importances, features)}
    feature_importances_norm = {fea: imp for imp, fea in zip(importances_norm, features)}
    # Sorts the dictionary
    feature_importances = {k: v for k, v in sorted(feature_importances.items(), key=lambda item: item[1], reverse = True)}
    feature_importances_norm= {k: v for k, v in sorted(feature_importances_norm.items(), key=lambda item: item[1], reverse = True)}
    # Prints the feature importances
    for k, v in feature_importances.items():
        print(f"{k} -> {v:.4f} (softmax = {feature_importances_norm[k]:.4f})")
```

```python
print_feature_importances_shap_values(shap_values, features[1:])
```

```python
shap.plots.bar(shap_values)
```

Here the features are ordered from the highest to the lowest effect on the prediction. It takes in account the absolute SHAP value, so it does not matter if the feature affects the prediction in a positive or negative way.

```python
shap.plots.beeswarm(shap_values)
# or shap.summary_plot(shap_values)
```

On the beeswarm the features are also ordered by their effect on prediction, but we can also see how higher and lower values of the feature will affect the result.

All the little dots on the plot represent a single observation. The horizontal axis represents the SHAP value, while the color of the point shows us if that observation has a higher or a lower value, when compared to other observations.

In this example, higher Weights and Heights have a negative impact on the prediction, while lower values have a positive impact.

```python
shap.summary_plot(shap_values, plot_type='violin')
```

### Analysis of local, instance-wise effects

We can use the following plots on single observations

```python
shap.plots.bar(shap_values[0])
```

This plot shows us what are the main features affecting the prediction of a single observation, and the magnitude of the SHAP value for each feature.

```python
shap.plots.waterfall(shap_values[0])
```

The waterfall plot has the same information, represented in a different manner. Here we can see how the sum of all the SHAP values equals the difference between the prediction f(x) and the expected value E[f(x)].

```python
shap.initjs() # This line is needed for the Force plot to be displayed.
shap.plots.force(shap_values[0])
```

The force plot is another way to see the effect each feature has on the prediction, for a given observation. In this plot the positive SHAP values are displayed on the left side and the negative on the right side, as if competing against each other. The highlighted value is the prediction for that observation.
